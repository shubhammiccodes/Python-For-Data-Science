## String formating in Python f- string

```python
name = "Shubham"
age = 30
```

```python
print("My name is {} and I am {} yaers old.".format(name, age)) 
```

## f-string:
- advance version of format() in String

```python
print(f"My name is {name} and I am {age} years old.")
```

```python
print(f"{"Python":<10}") # left align
print(f"{"Python":>10}") # right align
print(f"{"Python":^10}") # center align
    
```
